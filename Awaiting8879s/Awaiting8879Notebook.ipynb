```python
eFileAwaiting_df['DateObject'] = pd.to_datetime(eFileAwaiting_df['Last Move Date'])
eFileAwaiting_df.head()
```

```python
# Now let's see if we can read in that email address list...
# That's going to be tricky, but we can convert names to emails using regex, since company emails are formulaic.
import warnings
warnings.filterwarnings('ignore')
eFileAwaiting_df['last_name'] = eFileAwaiting_df['Responsible Party'].str.extract('(\w+),\s\w+\s\w\.')
eFileAwaiting_df['first_name'] = eFileAwaiting_df['Responsible Party'].str.extract('\w+,\s(\w+)\s\w\.')
eFileAwaiting_df['email'] = eFileAwaiting_df['first_name'] + '_' + eFileAwaiting_df['last_name'] + '@jrbt.com'
eFileAwaiting_df.drop(columns=['first_name','last_name'],inplace=True)
eFileAwaiting_df
```

```python
the24th = pd.to_datetime('07/24/2024')
long_delay_df = eFileAwaiting_df[eFileAwaiting_df['DateObject'] <= the24th]

long_delay_df.tail(10)
```
