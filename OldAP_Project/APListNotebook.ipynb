```python
import pandas as pd

AP_df = pd.read_excel('H:\Downloads\Search_Results6_8_202413_18_48 GMT-0500 (Central Daylight Time).xlsx',dtype={'Number':str})
AP_df
```

```python
Pickup_df = AP_df[['Client/Entity Name','Task Type','Number','Last Move Date','Responsible Party']]

Pickup_df
```

```python
import re
Pickup_df['Task Type'] = Pickup_df['Task Type'].str.extract('(1120\sS|\d{3,4})')
# The order of the two exps in () is significant because 'extract' returns the first one it finds
Pickup_df
```

```python
from datetime import datetime as dt
from datetime import timedelta

today = dt.today()
today
```

```python
one_week = timedelta(days=7)
two_weeks = timedelta(days=14)
print(f'Two weeks ago: {today - two_weeks}')
print(f'One week ago: {today - one_week}')
print(f'Today: {today}')
```

```python
Pickup_df['DateObject'] = pd.to_datetime(Pickup_df['Last Move Date'])
Pickup_df.head()
```

```python
# Now let's see if we can read in that email address list...
# That's going to be tricky, but we can convert names to emails using regex, since company emails are formulaic.
import warnings
warnings.filterwarnings('ignore')
Pickup_df['last_name'] = Pickup_df['Responsible Party'].str.extract('(\w+),\s\w+\s\w\.')
Pickup_df['first_name'] = Pickup_df['Responsible Party'].str.extract('\w+,\s(\w+)\s\w\.')
Pickup_df['email'] = Pickup_df['first_name'] + '_' + Pickup_df['last_name'] + '@jrbt.com'
Pickup_df.drop(columns=['first_name','last_name'],inplace=True)
Pickup_df
```

```python
two_weeks_df = Pickup_df[Pickup_df['DateObject'] <= (today - two_weeks)]

one_week_df = Pickup_df[Pickup_df['DateObject'] <= (today - one_week)]
one_week_df = one_week_df[one_week_df['DateObject'] > (today - two_weeks)]
two_weeks_df
```
